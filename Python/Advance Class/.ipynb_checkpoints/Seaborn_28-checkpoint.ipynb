```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_data = pd.read_csv('Datasets/Pokemon.csv') 
```

```python
sns.lmplot( x = 'Speed Point',
           y ='Defense Point',
           data = pokemon_data)
plt.show()
```

```python
sns.lmplot( x = 'Attack Point',
           y ='Defense Point',
           data = pokemon_data)
plt.show()
```

```python
sns.lmplot( x = 'Attack Point',
           y ='Defense Point',
           data = pokemon_data,
           size = 3,
           aspect = 2)
plt.show()
```

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data,
           x_estimator = np.median)
plt.show()
```

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data)
plt.show()
```

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data,
          col = 'Type 1')
plt.show()
```

```python
sns.lmplot(x ='Attack Point',
           y ='Defense Point',
           data = pokemon_data,
           col = 'Type 1',
           hue = 'Type 2')
plt.show()
```

```python
sns.regplot(x ='Attack Point',
            y ='Defense Point',
            data = pokemon_data)
plt.show()
```

```python
sns.regplot(x= pokemon_data['Attack Point'],
            y= pokemon_data['Defense Point'])
plt.show()
```

```python
plt.subplots(figsize=(7,6))
sns.regplot(x = 'Attack Point',
            y ='Defense Point',
            data = pokemon_data)
plt.show()
```

```python
sns.jointplot(x="Attack Point",
              y="Defense Point",
              data=pokemon_data,
             xlim={0,150},
              ylim={0,150},
              kind='reg')
```

```python
sns.pairplot(pokemon_data,
            size=5,aspect=0.9,
            x_vars=['Attack Point','Speed Point','Defense Point'],
            y_vars=['Total Point','Attack Point', 'Speed Point'],
            kind='reg')
plt.show()
```
