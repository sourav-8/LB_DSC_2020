# Different Visualization techniques in Seaborn

```python
# Univariate - single variable 
# Bivariate - two variabes - Scatter plot
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pokemon_data = pd.read_csv('Pokemon.csv')
```

```python
sns.distplot(pokemon_data['Attack'])
plt.show()
```

```python
sns.distplot(pokemon_data['Attack'], color = 'black')
plt.show()
```

```python
hist = sns.distplot(pokemon_data['Attack'], kde = False)
hist.set_title('Attack capability')
hist.set_xlabel('Attack Point')
hist.set_ylabel('Frequency')
plt.show()
```

```python
hist = sns.distplot(pokemon_data['Attack'], hist = False)
hist.set_title('Kernel Density plot')
hist.set_xlabel('Attack')
hist.set_ylabel('Unit Probability')
plt.show()
```

```python
hist = sns.distplot(pokemon_data['Attack'], rug = True)
hist.set_title('Attack capability with Density and Rug plot')
hist.set_xlabel('Attack')
plt.show()
```

```python
sns.rugplot(pokemon_data['Attack'],
            height = 0.5,
            axis = 'x')
plt.show()
```

```python
sns.distplot(pokemon_data['Attack'],
             bins = 40,
             kde = False,
             rug = True)
plt.show()
```

```python
sns.distplot(pokemon_data['Attack'],
             bins = 75,
             kde = False,
             rug = True)
plt.show()
```

```python
sns.distplot(pokemon_data['Attack'],
             bins = 15,
             kde = False,
             rug = True)
plt.show()
```

```python
sns.distplot(pokemon_data['Attack'],
             bins = 40,
             hist = False,
             rug = True)
plt.show()
```

```python
sns.set(color_codes = True)
sns.kdeplot(pokemon_data['Attack'],
            shade = True,
            color = 'g')
plt.show()
```

```python
hist = plt.subplots(figsize=(20,5))
sns.kdeplot(pokemon_data['Attack'])
sns.kdeplot(pokemon_data['Attack'],bw = 0.7,label ='bw:0.7')
sns.kdeplot(pokemon_data['Attack'],bw = 6,label ='bw:6')
sns.kdeplot(pokemon_data['Attack'],bw = 30,label ='bw:30')
sns.kdeplot(pokemon_data['Attack'],bw = 110,label ='bw:110')
plt.show()
```

```python
sns.jointplot(x ='Attack',
              y ='Defense',
              data = pokemon_data)
plt.show()
```

```python
sns.jointplot(x = 'Attack',
              y ='Defense',
              data = pokemon_data,
              xlim = {0,450},
              ylim = {0,200})
plt.show()
```

```python
with sns.axes_style('white'):
    sns.jointplot(x ='Attack',
                  y ='Defense',
                  data = pokemon_data,
                  kind ='hex',
                  color ='b')
```

```python
with sns.axes_style('white'):
    sns.jointplot(x ='Attack',
                  y ='Defense',
                  data = pokemon_data,
                  kind ='kde')
```

```python
sns.kdeplot(pokemon_data['Attack'],
            pokemon_data['Defense'],
            color = 'r',
            )
plt.show()
```

```python
sns.kdeplot(pokemon_data['Attack'],
            pokemon_data['Defense'],
            clip = (0,100))
plt.show()
```

```python
sns.kdeplot(pokemon_data['Attack'],
            pokemon_data['Defense'],)
plt.xlim(0,750)
plt.ylim(0,500)
```

```python
sns.kdeplot(pokemon_data['Attack'],
            pokemon_data['Defense'],)
sns.rugplot(pokemon_data['Attack'],
            color = 'r',
            height = .05)
sns.rugplot(pokemon_data['Defense'],
            color = 'b',
            vertical = True,
            height = .05)
plt.show()
```

```python
cormat = pokemon_data.corr()
plt.subplots(figsize = (7,7))
sns.heatmap(cormat, vmax = .8,
            square = True,
            annot = True, 
            fmt ='.2f', 
            cmap = 'Blues')
plt.show()
```
