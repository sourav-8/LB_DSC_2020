```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_data = pd.read_csv('Pokemon.csv') 
pokemon_data.head(2)
```

```python
sns.lmplot( x = 'Speed',
           y ='Defense',
           data = pokemon_data)
plt.show()
```

```python
sns.lmplot( x = 'Attack',
           y ='Defense',
           data = pokemon_data)
plt.show()
```

```python
sns.lmplot( x = 'Attack',
           y ='Defense',
           data = pokemon_data,
           size = 5,
           aspect = 2)
plt.show()
```

```python
sns.lmplot(x ='Attack',
           y ='Defense',
           data = pokemon_data,
           x_estimator = np.median)
plt.show()
```

```python
sns.lmplot(x ='Attack',
           y ='Defense',
           data = pokemon_data)
plt.show()
```

```python
sns.lmplot(x ='Attack',
           y ='Defense',
           data = pokemon_data,
          col = 'Type 1')
plt.show()
```

```python
sns.regplot(x ='Attack',
            y ='Defense',
            data = pokemon_data)
plt.show()
```

```python
sns.regplot(x= pokemon_data['Attack'],
            y= pokemon_data['Defense'])
plt.show()
```

```python
plt.subplots(figsize=(7,6))
sns.regplot(x = 'Attack',
            y ='Defense',
            data = pokemon_data)
plt.show()
```

```python
sns.jointplot(x="Attack",
              y="Defense",
              data=pokemon_data,
             xlim={0,150},
              ylim={0,150},
              kind='reg')
```

```python
sns.pairplot(pokemon_data,
            size=5,aspect=0.9,
            x_vars=['Attack','Speed','Defense'],
            y_vars=['Total','Attack', 'Speed'],
            kind='reg')
plt.show()
```
