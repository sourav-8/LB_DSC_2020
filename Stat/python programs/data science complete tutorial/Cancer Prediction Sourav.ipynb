# Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
trainX,testX,trainY,testY = train_test_split(cancer_data.data, cancer_data.target,test_size=0.30,random_state=0)
```

```python
lr = LogisticRegression(class_weight='balanced')
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from scipy.stats import randint
lr = LogisticRegression()
param_dist = {  "solver": ['liblinear','saga'],
                "random_state": randint(2, 12),
              "max_iter": randint(2500,3000)}
#               "tol": randint(.0001,.0002)}
param_dist
```

```python
n_iter_search = 20
random_search = RandomizedSearchCV(lr, param_distributions=param_dist,
                                   n_iter=n_iter_search, cv=5)
X=cancer_data.data
y=cancer_data.target


random_search.fit(X, y)
```

```python
random_search.best_estimator_
```

```python
random_search.best_score_
```

```python
lr = random_search.best_estimator_
lr.fit(trainX,trainY)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,accuracy_score
pred_lr = lr.predict(testX)
print(classification_report(pred_lr, testY))
```

```python
score = lr.score(testX, testY)
print(score)
```

```python
from sklearn.model_selection import cross_val_score
scores_lr = cross_val_score(lr, cancer_data.data, cancer_data.target, cv =5)
scores_lr
```

```python
scores_lr.mean()
```

```python
from sklearn.metrics import classification_report

from sklearn import metrics

cm_lr = metrics.confusion_matrix(testY, pred_lr)
# print(cm)
plt.figure(figsize=(9,9))
sns.heatmap(cm_lr, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'YlGnBu');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(scores_lr.mean())
plt.title(all_sample_title, size = 15);
plt.savefig('toy_Digits_ConfusionSeabornCodementor.png')
```
